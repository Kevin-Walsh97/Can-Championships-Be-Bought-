```python
# Imports
import pandas as pd
import numpy as np
import datetime as dt
from pathlib import Path
from scipy.stats import ttest_ind
%matplotlib inline
```

# Read in the Required CSV files

```python
# Read in the Required Dataframes
# height_salaries dataframe
height_salaries_csv = Path('Cleansed_csv/height_salaries.csv')
# Read csv into a pandas Dataframe
height_salaries = pd.read_csv(height_salaries_csv)
height_salaries.head()
```

```python
# Read in the Required Dataframes
# salary_record dataframe
salary_record_csv = Path('Cleansed_csv/salary_record.csv')
# Read csv into a pandas Dataframe
salary_record = pd.read_csv(salary_record_csv)
salary_record.head()
```

```python
# Read in the Required Dataframes
# salary_game_stats dataframe
salary_game_stats_csv = Path('Cleansed_csv/salary_game_stats.csv')
# Read csv into a pandas Dataframe
salary_game_stats = pd.read_csv(salary_game_stats_csv)
salary_game_stats.head()
```

```python
# Read in the Required Dataframes
# height_salaries dataframe
opponent_points_salary_csv = Path('Cleansed_csv/opponent_points_salary.csv')
# Read csv into a pandas Dataframe
opponent_points_salary = pd.read_csv(opponent_points_salary_csv)
opponent_points_salary.head()
```

# Perform T-Test Analysis

```python
# Correlation with Salary and Player Height

# Group Dataframe by above and below the avg team hight
avg_height = height_salaries['player_height'].mean()
above_avg = height_salaries[(height_salaries['player_height'] >= avg_height)]
below_avg = height_salaries[(height_salaries['player_height'] < avg_height)]

# Perform the T-Test
ttest_ind(above_avg["Salary w/ Inflation"],below_avg["Salary w/ Inflation"])
```

```python
# Correlation with Salary and Wins

# Group DataFrame by above and below the avg number of wins
avg_wins = salary_record['W'].mean()
above_avg = salary_record[(salary_record['W'] >= avg_wins)]
below_avg = salary_record[(salary_record['W'] < avg_wins)]

# Perform the T-Test
ttest_ind(above_avg["Salary w/ Inflation"],below_avg["Salary w/ Inflation"])
```

```python
# Correlation with Salary and Points Scored

# Group DataFrame by above and below the avg points scored
avg_points = salary_game_stats['PTS'].mean()
above_avg = salary_game_stats[(salary_game_stats['PTS'] >= avg_points)]
below_avg = salary_game_stats[(salary_game_stats['PTS'] < avg_points)]

# Perform the T-Test
ttest_ind(above_avg["Salary w/ Inflation"],below_avg["Salary w/ Inflation"])
```

```python
# Correlation with Salary and Opponent Points Scored

# Group DataFrame by above and below the avg points allowed
avg_points_allowed = opponent_points_salary['OPP_PTS'].mean()
above_avg = opponent_points_salary[(opponent_points_salary['OPP_PTS'] >= avg_points_allowed)]
below_avg = opponent_points_salary[(opponent_points_salary['OPP_PTS'] < avg_points_allowed)]

# Perform the T-Test
ttest_ind(above_avg["Salary w/ Inflation"],below_avg["Salary w/ Inflation"])
```
