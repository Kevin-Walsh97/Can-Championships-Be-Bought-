```python
### SCHEDULE
from sportsreference.nba.schedule import Schedule
mil2019 = Schedule('MIL', year = '2019')
mil2019.dataframe.head()
```

```python
### ROSTERS BY YEAR
from sportsreference.nba.roster import Roster
gsw2007 = Roster('GSW', year = '2007')
for player in gsw2007.players:
    print(player.name, ':', player.player_id)
```

```python
### POINTS SCORED BY TEAM BY YEAR
from sportsreference.nba.teams import Teams
teams2020 = Teams(year = '2010')
for team in teams2020:
    print(team.name, 'Points Scored:', team.points)
```

```python
### ABBREVIATIONS
from sportsreference.nba.teams import Teams
teams2020 = Teams(year = '2020')
for team in teams2020:
    print(team.name, 'Abbrev:', team.abbreviation)
```
